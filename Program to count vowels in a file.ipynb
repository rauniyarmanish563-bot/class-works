```python
# Program to count vowels in a file

# Open the file in read mode
file = open("firstfile.txt", "r")

# Read the content of the file
data = file.read()

# Initialize vowel count
vowels = "aeiouAEIOU"
count = 0

# Count vowels
for char in data:
    if char in vowels:
        count += 1

# Display result
print("Number of vowels in the file:", count)

# Close the file
file.close()
```
