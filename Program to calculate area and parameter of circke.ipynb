```python
# circle_calculation_advanced.py

import math   # Importing math module to use the value of pi

try:
    # Taking radius input from the user
    radius = float(input("Enter radius of the circle: "))

    # Checking if radius is negative
    if radius < 0:
        # Manually raising an error for negative radius
        raise ValueError("Radius cannot be negative!")

except ValueError as e:
    # Handling invalid input or negative value error
    print("Error:", e)

else:
    # This block executes only if no exception occurs
    # Calculating area of the circle (πr²)
    area = math.pi * radius * radius

    # Calculating perimeter (circumference) of the circle (2πr)
    perimeter = 2 * math.pi * radius

    # Displaying the result
    print("\n--- Result ---")
    print("Area of circle:", area)
    print("Perimeter of circle:", perimeter)

finally:
    # This block always executes (whether error occurs or not)
    print("\nProgram executed successfully (end of program).")
```
