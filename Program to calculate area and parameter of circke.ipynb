```python
# circle_calculation.py

import math   # math module use kar rahe hain pi ki value ke liye

while True:
    try:
        # user se radius input lena
        radius = float(input("Enter radius of the circle: "))

        # check karna ki radius negative na ho
        if radius < 0:
            print("Error: Radius cannot be negative!\n")
            continue   # dobara input lene ke liye loop continue

        # area calculate karna (πr²)
        area = math.pi * radius * radius

        # perimeter (circumference) calculate karna (2πr)
        perimeter = 2 * math.pi * radius

        # result print karna
        print("\n--- Result ---")
        print("Area of circle:", area)
        print("Perimeter of circle:", perimeter)

        break   # correct input milne par loop stop

    except ValueError:
        # agar user galat input deta hai (text ya invalid number)
        print("Error: Please enter a valid numeric value!\n")
```
