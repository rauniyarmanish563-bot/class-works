```python
# Program for writting data into the file
#opening file for write operation
filev = open("firstfile.text","w")
# writting file sentences into the file
print("Enter any five sentences")
for x in range(5):
  #input of data from user
  sentence = input()
  # writting sentence into the liune
  filev.write(sentence)
  print("_________________________")
  #----------------------------------------------------
  print("Data sucessfully written")
  # closing the file
  filev.close()
```
